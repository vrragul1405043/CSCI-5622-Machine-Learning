```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation
from tensorflow.keras.optimizers import SGD
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, ExtraTreesClassifier
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score
from datetime import datetime
import pickle
from xgboost import XGBClassifier
from sklearn.multiclass import OneVsRestClassifier
from tensorflow.keras.utils import to_categorical
```

```python
trainFeaturesPath = '/content/drive/MyDrive/PS4/train-features.csv'
trainOutputPath = '/content/drive/MyDrive/PS4/train-output.csv'
testFeaturesPath = '/content/drive/MyDrive/PS4/test-features.csv'
finalPath = '/content/drive/MyDrive/PS4/submission.csv'
```

```python
trainFeatures = pd.read_csv(trainFeaturesPath)
trainOutput = pd.read_csv(trainOutputPath)
```

```python
trainFeatures.shape
```

```python
trainFeatures = trainFeatures.drop('Id', axis=1)
trainFeatures = trainFeatures.drop('feature_28', axis = 1)
```

```python
standardScaler = StandardScaler()
```

```python
numerical_features = ['feature_0', 'feature_1', 'feature_2', 'feature_3', 'feature_4', 'feature_5', 'feature_6', 'feature_7', 'feature_8', 'feature_9']
```

```python
standardScaler = StandardScaler()
```

```python
trainFeatures[numerical_features] = standardScaler.fit_transform(trainFeatures[numerical_features])
```

```python
# X_train, X_test, y_train, y_test = train_test_split(trainFeatures, trainOutput['Covtype'].values, test_size=0.20, random_state=42)
```

```python
# temp_y_train = y_train - 1
```

```python
oversampled_train_features = trainFeatures.append(trainFeatures)
oversampled_train_labels = np.concatenate((trainOutput['Covtype'].values, trainOutput['Covtype'].values),axis=0) - 1
```

```python
y_train_encoded = to_categorical(oversampled_train_labels, 7)
num_input_features = (53, )
```

```python
def build_model1():
  model = Sequential()
  model.add(Dense(350, input_shape=num_input_features, activation='relu'))
  model.add(Dropout(0.1))
  model.add(Dense(100, activation='swish'))
  model.add(Dropout(0.1))
  model.add(Dense(75, activation='swish'))
  model.add(Dropout(0.1))
  model.add(Dense(50, activation='swish'))
  model.add(Dropout(0.1))
  model.add(Dense(25, activation='swish'))
  model.add(Dropout(0.1))
  model.add(Dense(15, activation='relu'))
  model.add(Dropout(0.1))
  model.add(Dense(7, activation='softmax'))
  model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
  return model
```

```python
def build_model2():
  model = Sequential()
  model.add(Dense(450, input_shape=num_input_features, activation='relu'))
  model.add(Dropout(0.1))
  model.add(Dense(250, input_shape=num_input_features, activation='swish'))
  model.add(Dropout(0.1))
  model.add(Dense(100, activation='swish'))
  model.add(Dropout(0.1))
  model.add(Dense(75, activation='swish'))
  model.add(Dropout(0.1))
  model.add(Dense(50, activation='swish'))
  model.add(Dropout(0.1))
  model.add(Dense(25, activation='swish'))
  model.add(Dropout(0.1))
  model.add(Dense(15, activation='relu'))
  model.add(Dropout(0.1))
  model.add(Dense(7, activation='softmax'))
  model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
  return model
```

```python
def build_model3():
  model = Sequential()
  model.add(Dense(450, input_shape=num_input_features, activation='relu'))
  model.add(Dropout(0.1))
  model.add(Dense(250, input_shape=num_input_features, activation='swish'))
  model.add(Dropout(0.1))
  model.add(Dense(100, activation='swish'))
  model.add(Dropout(0.1))
  model.add(Dense(75, activation='swish'))
  model.add(Dropout(0.1))
  model.add(Dense(65, activation='swish'))
  model.add(Dropout(0.1))
  model.add(Dense(50, activation='swish'))
  model.add(Dropout(0.1))
  model.add(Dense(25, activation='swish'))
  model.add(Dropout(0.1))
  model.add(Dense(15, activation='relu'))
  model.add(Dropout(0.1))
  model.add(Dense(7, activation='softmax'))
  model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
  return model
```

```python
model_1 = build_model1()
model_2 = build_model2()
model_3 = build_model3()

models = []
models.append(model_1)
models.append(model_2)
models.append(model_3)
```

```python
trained_models = []
for model in models:
  model.fit(oversampled_train_features , y_train_encoded , batch_size=60, epochs=130)
  trained_models.append(model)
```

```python
test_df = pd.read_csv(testFeaturesPath)
test_df = test_df.drop('feature_28', axis = 1)
test_df[numerical_features] = standardScaler.transform(test_df[numerical_features])
print(len(test_df))
```

```python
models_prediction_array = np.zeros((len(test_df),3), dtype=int)
for idx,trained_model in enumerate(trained_models):
  y_predicted = trained_model.predict(test_df.drop(columns=["Id"]))
  y_pred = np.argmax(y_predicted, axis=1)
  y_pred = y_pred + 1
  models_prediction_array[:,idx] = y_pred
```

```python
print(models_prediction_array.shape)
```

```python
final_predictions = []
for idx, row in enumerate(models_prediction_array):
  final_predictions.append(np.argmax(np.bincount(row)))
```

```python
result_df = pd.DataFrame(list(zip(test_df.loc[:, "Id"], final_predictions )), columns=["Id", "Covtype"])
result_df.to_csv(finalPath, index=False)
```
